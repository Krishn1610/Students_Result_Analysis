```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("students_scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop  unnamed column

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

# Change weekly study hours coloumn

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

# Gender Distribution

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data=df, x="Gender", hue="Gender", palette=["#1f77b4", "#ff7f0e"], legend=False)
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
# from the above chart we have analysed that: 
# the number of females in data is more than the number of males
```

# Parent's Education and Student's Score

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore": 'mean', "WritingScore": 'mean'})
print(gb)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

```python
#from the above chart we have concluded that the education of the parents have good impact on their scores
```

# Parent's Marital Status and Student's Score

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore": 'mean', "WritingScore": 'mean'})
print(gb1)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent's Marital Status and Student's Score")
plt.show()
```

```python
#from the above chart we have concluded that there is no/negligible impact on the student's scores due to their parent's marital status
```

# Ethnic Group Distribution 

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()
l = ["group A","group B","group C","group D", "group E"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
print(mlist)
plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()

```
